```python
x = data_df.drop(["DEATH_EVENT"], axis=1)
y = data_df["DEATH_EVENT"]
```

```python
col_name =list(x.columns)
s_scaler = preprocessing.StandardScaler()
X_scaled = s_scaler.fit_transform(x)
X_scaled = pd.DataFrame(X_scaled, columns = col_name)
```

```python
X_scaled.describe().T
```

```python
plt.figure(figsize=(20,10))
sns.boxenplot(data=X_scaled)
```

```python
X_train ,X_tests , y_train, y_test = train_test_split(X_scaled,y,test_size=0.3)
```

```python
model1=svm.SVC()
```

```python
model1.fit(X_train,y_train)
```

```python
y_pred = model1.predict(X_tests)
```

```python
y_pred
```

```python
y_test
```

```python
print(classification_report(y_test,y_pred))
```

```python
early_stopping = callbacks.EarlyStopping(
    min_delta = 0.001 , patience = 20, restore_best_weights = True)

model1 = Sequential()

model1.add(Dense(units = 16,kernel_initializer = 'uniform', activation = 'relu', input_dim = 12))
model1.add(Dense(units = 8,kernel_initializer = 'uniform', activation = 'relu'))
model1.add(Dropout(0.25))
model1.add(Dense(units = 8,kernel_initializer = 'uniform', activation = 'relu'))
model1.add(Dropout(0.5))
model1.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
model1.compile(optimizer = 'adam', loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
model1.summary()
```

```python
history = model1.fit(X_train, y_train, batch_size = 25 , epochs = 100, callbacks=[early_stopping],
                   validation_split = 0.25)
```

```python
history_df = pd.DataFrame(history.history)

plt.plot(history_df.loc[:, ['loss']] , label = "Training loss")
plt.plot(history_df.loc[: , ['val_loss']] ,label = "Validation loss")
plt.legend()
plt.show
```

```python

plt.plot(history_df.loc[:, ['accuracy']] , label = "Training accuracy")
plt.plot(history_df.loc[: , ['val_loss']] ,label = "Validation accuracy")
plt.legend()
plt.show
```

```python
y_pred = model1.predict(X_tests)
```

```python
y_pred = (y_pred > 0.5)
```

```python
print(classification_report(y_test, y_pred))
```
